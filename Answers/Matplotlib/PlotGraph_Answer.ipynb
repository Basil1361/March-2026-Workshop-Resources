```python
# Solution 3:
import matplotlib.pyplot as plt
import numpy as np
# I can going to use fig, ax as solution:

# equation 1:
# I can going to use linspace for the straight line:

def straight_line(x):
    return (x+10)

x_values = np.linspace(-10,10,100)
y_values = straight_line(x_values)
fig,ax = plt.subplots()
ax.plot(x_values,y_values,color = 'orange', linewidth = 2, label = "y = x + 10")

# equation 2:
# I am going to use linspace again
def curved_line(x):
    return (x**2)

x_values2 = np.linspace(-10,10,100)
y_values2 = curved_line(x_values2)
ax.plot(x_values2,y_values2, color = '#1f77b4', linewidth = 1.5, label = "y = x ^ 2")


# UI
ax.grid(True, alpha = 0.5, linestyle = "--")
ax.set_title("Graph of x^2 and x + 10")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()

# Additional Comments: 
'''
# Intersection, set condition:
# Note: we cannot set y_values = y_values2 because it's in a list!
# We can take advice from https://stackoverflow.com/questions/28766692/how-to-find-the-intersection-of-two-graphs

idx = np.argwhere(np.diff(np.sign(y_values2 - y_values))).flatten()
ax.plot(x_values[idx], y_values[idx], 'ro')


for i in idx:
    ax.annotate(
        text=f"({x_values[i]:.2f}, {y_values[i]:.2f})",
        xy=(x_values[i], y_values[i]-5),
        fontsize=8,
        color="green"
    )
'''
```

```python
# Solution 4

import matplotlib.pyplot as plt
import numpy as np

# 1. Define the function
def my_function(x):
    return np.sin(x)

# 2. Generate x-values (more points for smoother curve)
x_values = np.linspace(0, np.pi*2,100)  # 100 points for smooth curve

# 3. Calculate y-values
y_values = my_function(x_values)

# 4. Find the minimum point
min_idx = np.argmin(y_values)  # Index of minimum y-value
x_min = x_values[min_idx]
y_min = y_values[min_idx]

max_idx = np.argmax(y_values)
x_max = x_values[max_idx]
y_max = y_values[max_idx]

# 5. Plot the function
plt.figure(num = 0, dpi = 100)
plt.plot(x_values, y_values, linestyle='-', label='sin(x)')

# Mark the minimum point
plt.plot(x_min, y_min, color = 'red', marker = 'o', markersize=5, label=f'Min point')
plt.plot(x_max, y_max, color = 'blue', marker = 'o', markersize=5, label=f'Max point')
plt.xticks([0, np.pi/2, np.pi, np.pi*1.5, 2*np.pi], ['0', '$\pi/2$','$\pi$', '3$\pi/2$','$2\pi$'])

# 6. Add labels and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph of sin(x)')
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='k', linestyle='-', linewidth=1)
plt.axvline(x=0, color='k', linestyle='-', linewidth=1)

plt.text(x_min-0.5, y_min-0.1, 'Min Point',fontsize=10)
plt.text(x_max-0.5, y_max+0.1, 'Max Point',fontsize=10)


# 7. Display the plot
plt.show()
```
